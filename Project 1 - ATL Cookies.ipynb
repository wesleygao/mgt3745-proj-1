```python
import datetime
import math
from prettytable import PrettyTable

#=========================================================================
# menu options in a nested dict
menu = {  'Delicious Duo' : {'Sugar' : 3, 'Chocolate' : 3, 'Peanut' : 0 }, 
               'Six Pack' : {'Sugar' : 2, 'Chocolate' : 2, 'Peanut' : 2 }, 
            'Sugar Snack' : {'Sugar' : 6, 'Chocolate' : 0, 'Peanut' : 0 },
         'Burdell Bundle' : {'Sugar' : 5, 'Chocolate' : 5, 'Peanut' : 5 } }

# individual prices for cookies in a dict
cookiePrices = {'Sugar' : 1.00, 'Chocolate' : 2.00, 'Peanut' : 3.00}

# current cookie inventory in a dict
cookieInv    = {'Sugar' : 50,   'Chocolate' : 35,   'Peanut' : 20}

# recorded past orders in a nested dict (in reality, would be in a DBMS)
'Legend: SS => Sugar Snack ; SP => Six Pack; DD => Delicious Duo; BB => Burdell Bundle'
'Content:  tranCode, orderQty, orderDateTime, tranCost' 
orderRec = { 'SS1' : {5, '2022-02-26 13:52', 30.00 }, 
             'SS2' : {2, '2022-02-26 15:13', 12.00 },
             'DD1' : {3, '2022-02-26 15:32', 27.00 },
             'SP1' : {1, '2022-02-26 15:43', 12.00 },
             'DD2' : {6, '2022-02-26 16:03', 42.00 },
             'BB1' : {2, '2022-02-26 16:12', 30.00 },
             'DD3' : {4, '2022-02-26 16:14', 36.00 }         
           }

# =============================================================

# do work here

def displayMenu():
    table = PrettyTable()
    table.add_column('Menu Items', ['Sugar Snack', 'Six Pack', 'Delicious Duo', 'Burdell Bundle'])
    table.add_column('Sugar Cookie', [6, 2, 3, 5])
    table.add_column('Chocolate Cookie', ['', 2, 3, 5])
    table.add_column('Peanut Cookie', ['', 2, '', 5])
    table.add_column('Menu Item Price', [getMenuPrice('Sugar Snack'), 
                                         getMenuPrice('Six Pack'),
                                         getMenuPrice('Delicious Duo'),
                                         getMenuPrice('Burdell Bundle')])
    print(table) 

    
def getMenuPrice(menuItem):
    menuItemPrice = 0
    for k1,dict2 in menu.items():
        if menuItem == k1:
            for k2,v2 in dict2.items():
                #print('Type of cookie:', k2, & 'Number of cookies:', v2)
                if k2 in cookiePrices:
                    price = v2 * cookiePrices[k2]
                    menuItemPrice += price
    return menuItemPrice 


def maxUnitsCanOrder(menuItem, orderQty):
    
    if menuItem == 'Sugar Snack':
        avail_list = []
        
        try:
            Sugar_Snack_Avail = cookieInv['Sugar'] / menu['Sugar Snack']['Sugar']
            Sugar_Snack_Avail = math.floor(Sugar_Snack_Avail)
            avail_list.append(Sugar_Snack_Avail)

        except ZeroDivsionError:
            avail_list.append(0)
        
        min_avail_list = min(avail_list)
        
        if orderQty > min_avail_list and min_avail_list != 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and can only cover ' + str(min_avail_list) + ' ' + menuItem, '. Are we good to go with this quantity?')
            decision = input(("Yes or No?: "))
            if decision == 'No':
                print('Well don\'t be a stranger! We\'ll expect to see you back sometime soon!')
            elif decision == "Yes":
                print('Super! Cookies coming up. I have to say, you are looking terrific today!')
                print('Your total cost for', min_avail_list, menuItem, 'boxes will be $'+ str(min_avail_list * getMenuPrice(menuItem)) + '. Time to pay up!')
                processOrder(menuItem, min_avail_list)
                
        elif orderQty <= min_avail_list:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('Yes we can do that! Hot cookies coming right up. Hang tight!')
            print('Your total cost for', orderQty, menuItem, 'boxes will be $'+ str(orderQty * getMenuPrice(menuItem)) + '. Time to pay up!')
            processOrder(menuItem, orderQty)
            
        elif min_avail_list == 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and there are no '+ menuItem + ' left. See you tomorrow!')
          
    
    elif menuItem == 'Six Pack':
        avail_list = []
        
        try:
            Six_Pack1_Avail = cookieInv['Sugar'] / menu['Six Pack']['Sugar']
            Six_Pack1_Avail = math.floor(Six_Pack1_Avail)
            avail_list.append(Six_Pack1_Avail)

            Six_Pack2_Avail = cookieInv['Chocolate'] / menu['Six Pack']['Chocolate']
            Six_Pack2_Avail = math.floor(Six_Pack2_Avail)
            avail_list.append(Six_Pack2_Avail)

            Six_Pack3_Avail = cookieInv['Peanut'] / menu['Six Pack']['Peanut']
            Six_Pack3_Avail = math.floor(Six_Pack3_Avail)
            avail_list.append(Six_Pack3_Avail)
        
        except ZeroDivsionError:
            avail_list.append(0)
        
        min_avail_list = min(avail_list)
        
        if orderQty > min_avail_list and min_avail_list != 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and can only cover ' + str(min_avail_list) + ' ' + menuItem, '. Are we good to go with this quantity?')
            decision = input(("Yes or No?: "))
            if decision == 'No':
                print('Well don\'t be a stranger! We\'ll expect to see you back sometime soon!')
            elif decision == "Yes":
                print('Super! Cookies coming up. I have to say, you are looking terrific today!')
                print('Your total cost for', min_avail_list, menuItem, 'boxes will be $'+ str(min_avail_list * getMenuPrice(menuItem)) + '. Time to pay up!')
                processOrder(menuItem, min_avail_list)
                
        elif orderQty <= min_avail_list:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('Yes we can do that! Hot cookies coming right up. Hang tight!')
            print('Your total cost for', orderQty, menuItem, 'boxes will be $'+ str(orderQty * getMenuPrice(menuItem)) + '. Time to pay up!')
            processOrder(menuItem, orderQty)
            
        elif min_avail_list == 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and there are no '+ menuItem + ' left. See you tomorrow!')
     
    
    elif menuItem == 'Delicious Duo':
        avail_list = []
        
        try:
            Delicious_Duo1_Avail = cookieInv['Sugar'] / menu['Delicious Duo']['Sugar']
            Delicious_Duo1_Avail = math.floor(Delicious_Duo1_Avail)
            avail_list.append(Delicious_Duo1_Avail)

            Delicious_Duo2_Avail = cookieInv['Chocolate'] / menu['Delicious Duo']['Chocolate']
            Delicious_Duo2_Avail = math.floor(Delicious_Duo2_Avail)
            avail_list.append(Delicious_Duo2_Avail)
        
        except ZeroDivsionError:
            avail_list.append(0)
        
        min_avail_list = min(avail_list)
        
        if orderQty > min_avail_list and min_avail_list != 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and can only cover ' + str(min_avail_list) + ' ' + menuItem, '. Are we good to go with this quantity?')
            decision = input(("Yes or No?: "))
            if decision == 'No':
                print('Well don\'t be a stranger! We\'ll expect to see you back sometime soon!')
            elif decision == "Yes":
                print('Super! Cookies coming up. I have to say, you are looking terrific today!')
                print('Your total cost for', min_avail_list, menuItem, 'boxes will be $'+ str(min_avail_list * getMenuPrice(menuItem)) + '. Time to pay up!')
                processOrder(menuItem, min_avail_list)
                
        elif orderQty <= min_avail_list:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('Yes we can do that! Hot cookies coming right up. Hang tight!')
            print('Your total cost for', orderQty, menuItem, 'boxes will be $'+ str(orderQty * getMenuPrice(menuItem)) + '. Time to pay up!')
            processOrder(menuItem, orderQty)
            
        elif min_avail_list == 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and there are no '+ menuItem + ' left. See you tomorrow!')
              
        
    elif menuItem == 'Burdell Bundle':
        avail_list = []
        
        try:
            Burdell_Bundle1_Avail = cookieInv['Sugar'] / menu['Burdell Bundle']['Sugar']
            Burdell_Bundle1_Avail = math.floor(Burdell_Bundle1_Avail)
            avail_list.append(Burdell_Bundle1_Avail)

            Burdell_Bundle2_Avail = cookieInv['Chocolate'] / menu['Burdell Bundle']['Chocolate']
            Burdell_Bundle2_Avail = math.floor(Burdell_Bundle2_Avail)
            avail_list.append(Burdell_Bundle2_Avail)

            Burdell_Bundle3_Avail = cookieInv['Peanut'] / menu['Burdell Bundle']['Peanut']
            Burdell_Bundle3_Avail = math.floor(Burdell_Bundle3_Avail)
            avail_list.append(Burdell_Bundle3_Avail)
        
        except ZeroDivsionError:
            avail_list.append(0)
        
        min_avail_list = min(avail_list)
        
        if orderQty > min_avail_list and min_avail_list != 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and can only cover ' + str(min_avail_list) + ' ' + menuItem, '. Are we good to go with this quantity?')
            decision = input(("Yes or No?: "))
            if decision == 'No':
                print('Well don\'t be a stranger! We\'ll expect to see you back sometime soon!')
            elif decision == "Yes":
                print('Super! Cookies coming up. I have to say, you are looking terrific today!')
                print('Your total cost for', min_avail_list, menuItem, 'boxes will be $'+ str(min_avail_list * getMenuPrice(menuItem)) + '. Time to pay up!')
                processOrder(menuItem, min_avail_list)
                
        elif orderQty <= min_avail_list:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('Yes we can do that! Hot cookies coming right up. Hang tight!')
            print('Your total cost for', orderQty, menuItem, 'boxes will be $'+ str(orderQty * getMenuPrice(menuItem)) + '. Time to pay up!')
            processOrder(menuItem, orderQty)
            
        elif min_avail_list == 0:
            print('You have chosen to order', orderQty, 'boxes of menu item:', menuItem)
            print('So sorry!! We\'ve had a run on cookies and there are no '+ menuItem + ' left. See you tomorrow!')
    

def decreaseInv(menuItem, orderQty):
    for k1,v1 in cookieInv.items():
        for k2,dict2 in menu.items():
            if menuItem == k2:
                for k3,v3 in dict2.items():
                    v3 = v3 * orderQty
                    if k3 == k1:
                        updatedQty = v1 - v3
                        cookieInv[k1] = updatedQty
            
            
def recordOrder(menuItem, orderQty, tranCost):
    # gets initials of menuItem
    itemInitials = ''
    for k1,v1 in menu.items():
        if k1 == menuItem:
            for char in k1:
                if char == char.upper() and char != ' ':
                    itemInitials = itemInitials + char
    
    # checks if key is in orderRec, if not, adds it with the trans information
    numCount = 1
    newOrder = itemInitials + str(numCount)
    for k2 in orderRec:
        if newOrder == k2:
            numCount += 1
            newOrder = itemInitials + str(numCount)

    orderRec[newOrder] = {orderQty, datetime.datetime.now().strftime('%Y-%m-%d %H:%M'), tranCost}

    
def processOrder(menuItem, orderQty):
    decreaseInv(menuItem, orderQty)
    tranCost = orderQty * getMenuPrice(menuItem) # gets transaction cost 
    recordOrder(menuItem, orderQty, tranCost)
    
    
    
while True:
    print()
    print("Welcome to Hot Byte Cookies!")
    print('Here\'s our yummy menu!')
    displayMenu()

    print("Let's take your order")
    print('>>From our menu, please enter the Menu Item you want to buy:')
    menuItem = input()
    print('>>Please enter the number of', menuItem, 'boxes you want to buy:')
    orderQty = int(input())
    
    maxUnitsCanOrder(menuItem, orderQty)

    print()
    print('Updated orderRec ==>', orderRec)
    print()
    print('Ending cookie inventory ==>', cookieInv)
    break # uncomment if needed for break
```
